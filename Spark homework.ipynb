```python
from pyspark import SparkFiles
from pyspark import SparkContext
import re

sc = SparkContext()
```

```python
lines = sc.textFile("bbc_*.csv")
```

```python
lines.collect()
```

```python
#correct number of lines = 2226 + 9 = 2235
lines.coalesce(1).count()
```

```python
words = lines.flatMap(lambda line : re.sub(r'\s+', ' ', re.sub(r',|!|\.|\?|;|\)|\(|the|to|of|and|are|in|s|for|is|that', ' ', line)).split(" "))
words = words.filter(lambda x: x != '')
```

```python
words.coalesce(1).count()
```

```python
words.collect()
```

```python
tuples = words.map(lambda word : (word, 1))
counts = tuples.reduceByKey(lambda a, b: (a + b))
```

```python
counts.coalesce(1).collect()[2][1]
```

```python
counts.coalesce(1).top(10, key=lambda x: x[1])
```
